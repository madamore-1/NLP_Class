```python
# Import libraries
from nltk.corpus import wordnet as wn
from nltk.corpus import sentiwordnet as swn

# Part 1: WordNet Exploration

# 1. Retrieve synsets, lemma names, definitions, and examples for the word "dog"
print("Synsets, Lemma Names, Definitions, and Examples for 'dog':")
for synset in wn.synsets('dog'):
    print(f"Synset: {synset}")
    print(f"Lemmas: {synset.lemma_names()}")
    print(f"Definition: {synset.definition()}")
    print(f"Examples: {synset.examples()}\n")

# 2. Get the hypernyms of the synset 'dog.n.01'
dog1 = wn.synset('dog.n.01')
hypernyms = dog1.hypernyms()
print(f"Hypernyms of dog.n.01: {hypernyms}")

# 3. Retrieve the hypernym paths for 'dog.n.01'
print("\nHypernym Paths for 'dog.n.01':")
paths = dog1.hypernym_paths()
for path in paths:
    print([synset.name() for synset in path])

# 4. Get part meronyms and member holonyms for 'dog.n.01'
part_meronyms = dog1.part_meronyms()
member_holonyms = dog1.member_holonyms()
print(f"\nPart Meronyms of 'dog.n.01': {part_meronyms}")
print(f"Member Holonyms of 'dog.n.01': {member_holonyms}")

# Part 2: Sentiment Analysis with SentiWordNet

# 5. Retrieve sentiment scores for the synset 'dog.n.01'
dogswn1 = swn.senti_synset('dog.n.01')
print("\nSentiment Scores for 'dog.n.01':")
print(f"Positive Score: {dogswn1.pos_score()}")
print(f"Negative Score: {dogswn1.neg_score()}")
print(f"Objective Score: {dogswn1.obj_score()}")
```
